```python
#fig, ax = plt.subplots()
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)

# Plot North-South component
ax.plot(df_CO56['Time (s)'], df_CO56['CO56-NS'], lw=4, c='red', label='CO56-NorthSouth')

# Plot East-West component
ax.plot(df_CO56['Time (s)'], df_CO56['CO56-EW'], lw=4, c='blue', label='CO56-EastWest')


# Add line for period and spectral accelaration
x_x=[1.512,1.512]
y_y=[0.0,0.24]

x_x2=[0,1.512]
y_y2=[0.240,0.240]

#ax.plot(x_x, y_y, lw=4, c='green')
#ax.plot(x_x2, y_y2, lw=4, c='green')

# Set axes limits
ax.set_xlim(0,5)
ax.set_ylim(0,0.5)

# Tick parameters
ax.tick_params(axis="x", labelsize=24)
ax.tick_params(axis="y", labelsize=24)

# Axes titles
ax.set_xlabel(r'Period $T_1$ (s)', fontsize = 24, labelpad=10)
ax.set_ylabel('Spectral acceleration (g)', fontsize = 24, labelpad=10)

# Grid lines
ax.grid(linewidth=1)

# Legend
ax.legend()
ax.legend(loc=1,prop={'size': 20})  #to make legend bold

plt.show()
```
